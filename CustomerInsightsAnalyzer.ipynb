```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('/content/ecommerce_customer_data.csv')
df
```

```python
df.info()
```

```python
df.groupby('Gender').sum()
```

```python
#people only purchase roughly half of the items they initially add to their shopping carts.
```

```python
px.histogram(df,x='Gender',y='Total_Purchases').show()
```

```python
#from the above plot its evident that the male purchase rate is much higher than females
```

```python
px.histogram(df,x='Gender',y='Product_Browsing_Time')
```

```python
#most people browsing through the product are men
```

```python
df.groupby('Device_Type').sum()
```

```python
px.histogram(df,x='Device_Type',y='Product_Browsing_Time')
```

```python
#from the above graph its clear that the most time spend in browsing is from mobiles
```

```python
df.groupby('Location').sum()
```

```python
px.histogram(df,x='Location',y='Total_Purchases')
```

```python
#from the above graph its clear that delhi has seen the highest number of purchases
```

```python
px.histogram(df,x='Location',y='Product_Browsing_Time')
```

```python
#from the above graph its clear that kolkata has the most browsing time
```

```python
px.histogram(df,x='Age',y='Product_Browsing_Time',nbins=18,color='Gender')
```

```python
#The women who have spent the most time browsing are in the age group of 18, whereas men are from the age group of 29
```

```python
px.histogram(df,x='Age',y='Total_Purchases',nbins=18,color='Gender')
```

```python
#The women who have purchased the most are in the age group of 18, whereas men are from the age group of 18 ,21,22.
```
